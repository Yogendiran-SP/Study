```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df = pd.read_csv("../CSV/HR_comma_sep.csv")
df
```

left vs number_project, average_montly_hours, last_evaluation, satisfaction_level, time_spend_company

```python
sns.barplot(x="Department",y="left",data=df,estimator=lambda x: sum(x)/len(x))
plt.title("Retention rate by Department")
plt.xlabel("Department")
plt.ylabel("Left")
plt.xticks(rotation=45)
```

```python
sns.barplot(x="salary",y="left",data=df,estimator=lambda x: sum(x)/len(x))
plt.title("Retention rate by Salary")
plt.xlabel("Salary")
plt.ylabel("Left")
plt.xticks(rotation=45)
```

```python
X = df[['number_project', 'average_montly_hours', 'last_evaluation', 'satisfaction_level', 'time_spend_company']]
X
```

```python
Y = df.left
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, Y_test)
```
