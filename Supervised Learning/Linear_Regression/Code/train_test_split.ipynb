```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../CSV/bmwprice.csv")
df
```

```python
X = df.drop(['Sell Price($)'],axis='columns')
X
```

```python
Y = df['Sell Price($)']
Y
```

```python
plt.xlabel("Mileage")
plt.ylabel("Sell Price($)")
plt.scatter(X.Mileage,Y,color='red',marker='+')
```

```python
plt.xlabel("Age(yrs)")
plt.ylabel("Sell Price($)")
plt.scatter(X['Age(yrs)'],Y,color='red',marker='+')
```
